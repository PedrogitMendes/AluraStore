### Importação dos dados



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja.head()
```

#1. Análise do faturamento


```python
faturamento_loja1 = loja['Preço'].sum()
faturamento_loja2 = loja2['Preço'].sum()
faturamento_loja3 = loja3['Preço'].sum()
faturamento_loja4 = loja4['Preço'].sum()

print(f"Faturamento total Loja 1: R$ {faturamento_loja1:.2f}")
print(f"Faturamento total Loja 2: R$ {faturamento_loja2:.2f}")
print(f"Faturamento total Loja 3: R$ {faturamento_loja3:.2f}")
print(f"Faturamento total Loja 4: R$ {faturamento_loja4:.2f}")
```

# 2. Vendas por Categoria


```python
vendas_por_categoria_loja1 = loja['Categoria do Produto'].value_counts()
vendas_por_categoria_loja2 = loja2['Categoria do Produto'].value_counts()
vendas_por_categoria_loja3 = loja3['Categoria do Produto'].value_counts()
vendas_por_categoria_loja4 = loja4['Categoria do Produto'].value_counts()

print("Vendas por Categoria - Loja 1:")
display(vendas_por_categoria_loja1)

print("\nVendas por Categoria - Loja 2:")
display(vendas_por_categoria_loja2)

print("\nVendas por Categoria - Loja 3:")
display(vendas_por_categoria_loja3)

print("\nVendas por Categoria - Loja 4:")
display(vendas_por_categoria_loja4)
```

# 3. Média de Avaliação das Lojas

```python
media_avaliacao_loja1 = loja['Avaliação da compra'].mean()
media_avaliacao_loja2 = loja2['Avaliação da compra'].mean()
media_avaliacao_loja3 = loja3['Avaliação da compra'].mean()
media_avaliacao_loja4 = loja4['Avaliação da compra'].mean()

print(f"Média de Avaliação Loja 1: {media_avaliacao_loja1:.2f}")
print(f"Média de Avaliação Loja 2: {media_avaliacao_loja2:.2f}")
print(f"Média de Avaliação Loja 3: {media_avaliacao_loja3:.2f}")
print(f"Média de Avaliação Loja 4: {media_avaliacao_loja4:.2f}")
```

# 4. Produtos Mais e Menos Vendidos

```python
produtos_mais_vendidos_loja1 = loja['Produto'].value_counts().head()
produtos_menos_vendidos_loja1 = loja['Produto'].value_counts().tail()

produtos_mais_vendidos_loja2 = loja2['Produto'].value_counts().head()
produtos_menos_vendidos_loja2 = loja2['Produto'].value_counts().tail()

produtos_mais_vendidos_loja3 = loja3['Produto'].value_counts().head()
produtos_menos_vendidos_loja3 = loja3['Produto'].value_counts().tail()

produtos_mais_vendidos_loja4 = loja4['Produto'].value_counts().head()
produtos_menos_vendidos_loja4 = loja4['Produto'].value_counts().tail()

print("Produtos Mais Vendidos - Loja 1:")
display(produtos_mais_vendidos_loja1)
print("\nProdutos Menos Vendidos - Loja 1:")
display(produtos_menos_vendidos_loja1)

print("\nProdutos Mais Vendidos - Loja 2:")
display(produtos_mais_vendidos_loja2)
print("\nProdutos Menos Vendidos - Loja 2:")
display(produtos_menos_vendidos_loja2)

print("\nProdutos Mais Vendidos - Loja 3:")
display(produtos_mais_vendidos_loja3)
print("\nProdutos Menos Vendidos - Loja 3:")
display(produtos_menos_vendidos_loja3)

print("\nProdutos Mais Vendidos - Loja 4:")
display(produtos_mais_vendidos_loja4)
print("\nProdutos Menos Vendidos - Loja 4:")
display(produtos_menos_vendidos_loja4)
```

# 5. Frete Médio por Loja


```python
frete_medio_loja1 = loja['Frete'].mean()
frete_medio_loja2 = loja2['Frete'].mean()
frete_medio_loja3 = loja3['Frete'].mean()
frete_medio_loja4 = loja4['Frete'].mean()

print(f"Frete médio Loja 1: R$ {frete_medio_loja1:.2f}")
print(f"Frete médio Loja 2: R$ {frete_medio_loja2:.2f}")
print(f"Frete médio Loja 3: R$ {frete_medio_loja3:.2f}")
print(f"Frete médio Loja 4: R$ {frete_medio_loja4:.2f}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Data for plotting
faturamento = [faturamento_loja1, faturamento_loja2, faturamento_loja3, faturamento_loja4]
lojas = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']
frete_medio = [frete_medio_loja1, frete_medio_loja2, frete_medio_loja3, frete_medio_loja4]
media_avaliacao = [media_avaliacao_loja1, media_avaliacao_loja2, media_avaliacao_loja3, media_avaliacao_loja4]

# Plotting Faturamento (Bar Plot)
plt.figure(figsize=(10, 6))
sns.barplot(x=lojas, y=faturamento)
plt.title('Faturamento Total por Loja')
plt.ylabel('Faturamento (R$)')
plt.xlabel('Loja')
plt.show()

# Plotting Frete Médio (Line Plot)
plt.figure(figsize=(10, 6))
sns.lineplot(x=lojas, y=frete_medio, marker='o')
plt.title('Frete Médio por Loja')
plt.ylabel('Frete Médio (R$)')
plt.xlabel('Loja')
plt.show()

# Plotting Média de Avaliação (Scatter Plot)
plt.figure(figsize=(10, 6))
ax = sns.scatterplot(x=lojas, y=media_avaliacao, s=100) # s is marker size
plt.title('Média de Avaliação por Loja')
plt.ylabel('Média de Avaliação')
plt.xlabel('Loja')
plt.ylim(3.8, 4.2) # Adjusted y-limit for better visualization of small differences

# Add labels to the points
for i, txt in enumerate(media_avaliacao):
    ax.annotate(f'{txt:.2f}', (lojas[i], media_avaliacao[i]), textcoords="offset points", xytext=(0,10), ha='center')

plt.show()
```
